```python
import matplotlib.pyplot as plt
import numpy as np

# 模擬月份資料
months = np.arange(1, 13)

# 模擬三個城市的月平均溫（攝氏）
taipei = [16, 17, 19, 23, 26, 28, 30, 30, 29, 26, 22, 18]
tokyo = [5, 6, 9, 14, 18, 21, 25, 26, 22, 17, 12, 8]
harbin = [-15, -10, 0, 10, 18, 23, 25, 22, 15, 5, -5, -12]

# 設定長條圖的寬度與位置
bar_width = 0.25
x = np.arange(len(months))

plt.figure(figsize=(10, 6))

# 每個城市的長條位置錯開
plt.bar(x - bar_width, taipei, width=bar_width, label='Taipei')
plt.bar(x, tokyo, width=bar_width, label='Tokyo')
plt.bar(x + bar_width, harbin, width=bar_width, label='Harbin')

# 標題與軸標籤
plt.title("Title", fontsize=14)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Temp. (°C)", fontsize=12)
plt.xticks(x, months)

# 顯示圖例與網格
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 顯示圖表
plt.tight_layout()
plt.show()
```
